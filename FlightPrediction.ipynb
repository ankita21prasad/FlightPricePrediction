# Flight Price Prediction (EDA + Feature Engineering)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel('Data_Train.xlsx')
test_df = pd.read_excel('Test_set.xlsx')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
#Merge train and test data
df = pd.concat([train_df, test_df])
df.head()
```

```python
df.info()
```

```python
#Feature Engineering
df['Date'] = df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month'] = df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year'] = df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
df.head()
```

```python
df = df.drop('Date_of_Journey', axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.isnull().sum()
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_Min'] = df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
df.head()
```

```python
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_Min'] = df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
df.info()
```

```python
# df.drop('Arrival_Time', axis =1 ,inplace = True)
df.drop('Dep_Time', axis =1 ,inplace = True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1})
```

```python
df.drop('Route', axis = 1, inplace = True)
```

```python
df.head()
```

```python
df['Duration_hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head()
```

```python
df[df['Duration_hour']=='5m']
```

```python
df.drop(6474,axis=0,inplace=True)
df.drop(2660,axis=0,inplace=True)
```

```python
df['Duration_min'].fillna(0, inplace = True)
```

```python
df['Duration_hour'] = df['Duration_hour'].astype(int)
df['Duration_min'] = df['Duration_min'].astype(int)
```

```python
df.head()
```

```python
df.drop('Duration', axis = 1, inplace = True)
```

```python
df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
df['Airline']=labelencoder.fit_transform(df['Airline'])
df['Source']=labelencoder.fit_transform(df['Source'])
df['Destination']=labelencoder.fit_transform(df['Destination'])
df['Additional_Info']=labelencoder.fit_transform(df['Additional_Info'])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
df['Airline'] = ohe.fit_transform(np.array(df['Airline']).reshape(-1, 1)).toarray()
df['Source'] = ohe.fit_transform(np.array(df['Source']).reshape(-1, 1)).toarray()
df['Destination'] = ohe.fit_transform(np.array(df['Destination']).reshape(-1, 1)).toarray()
df['Additional_Info'] = ohe.fit_transform(np.array(df['Additional_Info']).reshape(-1, 1)).toarray()
```
